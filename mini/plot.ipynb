# Setup, loading and plotting functions

```python
import matplotlib.pyplot as plt
import os
import glob
import numpy as np
import deepdish as dd
import scipy
import matplotlib.patches as mpatches
from mpl_toolkits.mplot3d import Axes3D
```

```python
def load(exp):
    step_names = glob.glob("results/{}/{}/*.npy".format(exp,'position'))
    step_list = np.sort([int(s.split(".npy")[0].split("/")[-1]) for s in step_names])
    
    position = []
    velocity = []
    for step in step_list:
        position.append(np.load("results/{}/{}/{}.npy".format(exp,'position', step)))
        velocity.append(np.load("results/{}/{}/{}.npy".format(exp,'velocity', step)))
    return np.vstack(position), np.vstack(velocity), step_list
```

```python
def hessian(exp):
    H = np.load("results/{}/hessian.npy".format(exp))
    W, V = np.linalg.eigh(H)
    return W, V
```

# Theory

```python
def trajectory(eigenvalue, initialization, lr, beta, lamb, time):

    gamma = (1-beta)/(lr*(1+beta))
    omega = np.sqrt(2/(lr*(1+beta))*(eigenvalue+lamb))

    if gamma < omega:
        freq = np.sqrt(omega**2 - gamma**2)
        a = np.exp(-gamma*time)*(np.cos(freq*time) + gamma/freq*np.sin(freq*time))
        b = -np.exp(-gamma*time)*omega**2/freq*np.sin(freq*time)

    if gamma == omega:
        a = np.exp(-gamma*time)*(1 + gamma*time)
        b = -np.exp(-gamma*time)* gamma**2 * time

    if gamma > omega:
        freq = np.sqrt(gamma**2 - omega**2)
        a = np.exp(-gamma*time)*(np.cosh(freq*time) + gamma/freq*np.sinh(freq*time))
        b = -np.exp(-gamma*time)*omega**2/freq*np.sinh(freq*time)

    return a * initialization, b * initialization
```

```python
p, v, steps = load('theory')
U = np.load("results/theory/eigenvector.npy")
S = np.load("results/theory/eigenvalues.npy")
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

k=np.shape(p)[0]
idx = 4

x = p.dot(U[:,idx])
y = -v.dot(U[:,idx])

time = np.linspace(0,max(steps),10000)
mu = np.mean(p[-100:,:], axis=0)
lr, beta, lamb = 1e-3, 0.9, 1e-4
a0 = (p[0,:] - mu).dot(U[:,idx])
a, b = trajectory(S[idx], a0, lr, beta, lamb, lr*time)
a += mu.dot(U[:,idx])

ax.plot(x, steps, y, lw=2, color='k')
ax.plot(a, time, b, lw=2, color='r')
ax.plot(x, steps, min(y)*np.ones(k), color='k', alpha=0.25)
ax.plot(min(x)*np.ones(k), steps, y, color='k', alpha=0.25)

ax.set_xlabel('Position')
ax.set_ylabel('Time')
ax.set_zlabel('Velocity')
ax.view_init(elev=25., azim=-35)

# ax.get_proj = lambda: np.dot(Axes3D.get_proj(ax), np.diag([0.5, 1, 0.5, 0.75]))
plt.tight_layout()
# plt.savefig('bias-oscillations-3d.png', format='png', bbox_inches='tight', transparent=True)
```
